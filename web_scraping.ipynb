```python
import requests
from bs4 import BeautifulSoup

# Send a GET request to the website
url = "https://www.flipkart.com/search?q=male%20clothes&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=off&as=off"
response = requests.get(url)

# Parse the HTML content of the page
soup = BeautifulSoup(response.content, "html.parser")

# Find all product containers
product_containers = soup.find_all("div", class_="single-product")

# Initialize lists to store product data
product_names = []
product_prices = []
product_image_urls = []

# Extract product data
for container in product_containers:
    # Product name
    name = container.find("h4", class_="title").text.strip()
    product_names.append(name)

    # Product price
    price = container.find("p", class_="price").text.strip()
    product_prices.append(price)

    # Product image URL
    image_url = container.find("img")["src"]
    product_image_urls.append(image_url)

# Display the scraped data
for i in range(len(product_names)):
    print("Product Name:", product_names[i])
    print("Price:", product_prices[i])
    print("Image URL:", product_image_urls[i])
    print()
```

```python
import requests
from bs4 import BeautifulSoup

# Send a GET request to the website
url = "https://www.flipkart.com/search?q=male%20clothes&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=off&as=off"
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3"
}
response = requests.get(url, headers=headers)

# Check if the request was successful
if response.status_code == 200:
    # Parse the HTML content of the page
    soup = BeautifulSoup(response.content, "html.parser")

    # Find all product containers
    product_containers = soup.find_all("div", class_="_1AtVbE")

    # Initialize lists to store product data
    product_names = []
    product_prices = []
    product_ratings = []

    # Extract product data
    for container in product_containers:
        # Product name
        name = container.find("a", class_="IRpwTa")
        if name:
            product_names.append(name.text.strip())
        else:
            product_names.append("Not available")

        # Product price
        price = container.find("div", class_="_30jeq3")
        if price:
            product_prices.append(price.text.strip())
        else:
            product_prices.append("Not available")

        # Product rating
        rating = container.find("div", class_="_3LWZlK")
        if rating:
            product_ratings.append(rating.text.strip())
        else:
            product_ratings.append("Not available")

    # Display the scraped data
    for i in range(len(product_names)):
        print("Product Name:", product_names[i])
        print("Price:", product_prices[i])
        print("Rating:", product_ratings[i])
        print()

else:
    print("Failed to retrieve data. Status code:", response.status_code)
```

```python
import requests
from bs4 import BeautifulSoup

# Send a GET request to the website
url = "https://www.amazon.com/s?k=male+clothes&crid=2Y475KRXQKS35&sprefix=male+cloth%2Caps%2C331&ref=nb_sb_noss_2"
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3"
}
response = requests.get(url, headers=headers)

# Check if the request was successful
if response.status_code == 200:
    # Parse the HTML content of the page
    soup = BeautifulSoup(response.content, "html.parser")

    # Find all product containers
    product_containers = soup.find_all("div", class_="s-result-item")

    # Initialize lists to store product data
    product_data = []

    # Extract product data
    for container in product_containers:
        # Product name
        name = container.find("span", class_="a-size-medium")
        if name:
            product_name = name.text.strip()
        else:
            product_name = "Not available"

        # Product price
        price = container.find("span", class_="a-offscreen")
        if price:
            product_price = price.text.strip()
        else:
            product_price = "Not available"

        # Product rating
        rating = container.find("span", class_="a-icon-alt")
        if rating:
            product_rating = rating.text.strip()
        else:
            product_rating = "Not available"

        # Product image URL
        image = container.find("img", class_="s-image")
        if image:
            image_url = image.get("src")
        else:
            image_url = "Not available"

        # Store product data in a dictionary
        product_data.append({
            "name": product_name,
            "price": product_price,
            "rating": product_rating,
            "image_url": image_url
        })

    # Display the scraped data
    for product in product_data:
        print("Product Name:", product["name"])
        print("Price:", product["price"])
        print("Rating:", product["rating"])
        print("Image URL:", product["image_url"])
        print()

else:
    print("Failed to retrieve data. Status code:", response.status_code)
```
